```python
import pandas as pd
import sys
import numpy as np
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
# Load the dataset
data = pd.read_csv('Applications_for_Machine_Learning_internship_edited.xlsx - Sheet1.csv')
data.insert(loc = 0,
          column = 'Intern_id',
          value = data.reset_index().index)
```

```python
data.head(2)
```

```python
data.isnull().sum()
```

```python
selected_features = data[['Python (out of 3)', 'Machine Learning (out of 3)',
                         'Natural Language Processing (NLP) (out of 3)', 'Deep Learning (out of 3)','Other skills','Intern_id']]
selected_features.isnull().sum()
```

```python
selected_features = selected_features.dropna()
selected_features.isnull().sum()
```

```python
# Select relevant features for clustering
numeric_features = selected_features[['Python (out of 3)', 'Machine Learning (out of 3)',
                         'Natural Language Processing (NLP) (out of 3)', 'Deep Learning (out of 3)']]

categorical_features = selected_features[['Other skills','Intern_id']]
categorical_features=categorical_features.drop(columns='Intern_id')                              
```

```python
numeric_features.isnull().sum()
```

```python
categorical_features.isnull().sum()
```

```python
# Scale the numerical features
scaler = StandardScaler()
numeric_features_scaled = scaler.fit_transform(numeric_features)

# One-hot encode the categorical features
encoder = OneHotEncoder(sparse=False)
categorical_features_encoded = encoder.fit_transform(categorical_features)

```

```python
numeric_features_scaled 
```

```python
categorical_features_encoded=pd.DataFrame(categorical_features_encoded)
```

```python
# Combine the scaled numeric features and encoded categorical features
features_combined = pd.concat([pd.DataFrame(numeric_features_scaled),
                               pd.DataFrame(categorical_features_encoded)],
                              axis=1)
features_combined
```

```python
# Determine the optimal number of clusters
# (Example using the elbow method)
inertia = []
k_values = range(1, 6)
for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(features_combined)
    inertia.append(kmeans.inertia_)

# Plot the elbow curve
import matplotlib.pyplot as plt
plt.plot(k_values, inertia, marker='o')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia')
plt.title('Elbow Curve')
plt.show()



# Apply K-means clustering
k = 4  # Set the chosen number of clusters
kmeans = KMeans(n_clusters=k, random_state=42)
kmeans.fit(features_combined)

# Assign cluster labels to the data
cluster_labels = kmeans.labels_
selected_features['Cluster'] = cluster_labels
```

```python
selected_features
```

```python
# Select the intern with the highest scores in relevant skills within each cluster
best_interns = selected_features.groupby('Cluster').apply(
    lambda x: x.iloc[x['Python (out of 3)'].idxmax()]
)
best_intern_candidates = best_interns[[ 'Intern_id','Python (out of 3)', 'Machine Learning (out of 3)',
                                       'Deep Learning (out of 3)', 'Other skills']]
best_intern_candidates
```

```python
# Select the intern with the highest overall score in relevant skills
best_intern = selected_features.iloc[selected_features[['Python (out of 3)', 'Machine Learning (out of 3)', 'Deep Learning (out of 3)']].sum(axis=1).idxmax()]

# Display the best intern
best_intern_candidates2 = best_intern[['Intern_id', 'Python (out of 3)', 'Machine Learning (out of 3)', 'Deep Learning (out of 3)', 'Other skills']]
best_intern_candidates2
```
